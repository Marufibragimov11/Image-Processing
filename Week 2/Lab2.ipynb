# Step 1

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('CAU.png')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
subtracted_image = cv2.subtract(image_RGB, 100)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-100)')
plt.axis('off')
plt.show()
```
