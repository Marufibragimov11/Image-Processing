# Part 1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gamma_correction(I_in, c, gamma):
    I_norm = I_in.astype('float32') / 255.0
    I_gamma = c * (I_norm ** gamma)
    I_out = np.clip(I_gamma * 255, 0, 255).astype('uint8')
    
    return I_out
```

```python
image = cv2.imread('1.png')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
c_value = 1
gamma_value = 0.5   # Change this depending on image (0.5 brightens, 2 darkens)

output = gamma_correction(image_gray, c_value, gamma_value)

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(image_gray, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f"Gamma Corrected ( = {gamma_value})")
plt.imshow(output, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Part 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def contrast_stretching(I_in, s_min, s_max):
    I_in = I_in.astype('float32')
    r_min = np.min(I_in)
    r_max = np.max(I_in)
    I_out = ((s_max - s_min) / (r_max - r_min)) * (I_in - r_min) + s_min
    
    return I_out
```

```python
image = cv2.imread('3.png')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

s_min = 0
s_max = 255

stretched = contrast_stretching(image_gray, s_min, s_max)

stretched_uint8 = np.clip(stretched, 0, 150).astype('uint8')

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(image_gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(stretched_uint8, cmap='gray')
plt.title("Contrast Stretched (0255)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
